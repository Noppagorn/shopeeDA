```python
def concat_all_str(list_str):
    temp = ""
    for x in list_str:
        temp += "" + str(int(x)) + "-"
    return temp[:len(temp) - 1]
def cut_nan(listArr):
    new_list = []
    for x in listArr:
        if str(x) != 'nan':
            new_list.append(x)
    return new_list
```

```python
tempArr = []
# allConect.ticket_trace = np.nan
for i, row in allConect.iterrows():
#     print(i)
    A_B_C = row.Id_A_B_C
    A_B_D = row.Id_A_B_D
    A_C_C = row.Id_A_C_C
    A_C_D = row.Id_A_C_D
    B_B_C = row.Id_B_B_C
    B_B_D = row.Id_B_B_D
    B_C_C = row.Id_B_C_C
    B_C_D = row.Id_B_C_D
    listArr = [A_B_C,A_B_D,A_C_C,A_C_D,B_B_C,B_B_D,B_C_C,B_C_D]
    listArr = cut_nan(listArr)
    listArr.sort()
    temp = concat_all_str(listArr)
    allConect.loc[i].ticket_trace = temp
#     allConect.set_value(i, 'ticket_trace', temp)
```

```python
def pluse_stringNan(lista):
    temp = 0
    for x in lista:
        if str(x) != 'nan':
            temp += x
    return str(temp)
```

```python

allConect["ticket_trace/contact"] = str(allConect.ticket_trace) + ", " + pluse_stringNan([allConect.Contacts_A_B_C,
                        allConect.Contacts_A_B_D ,allConect.Contacts_A_C_C , allConect.Contacts_A_C_D , 
                        allConect.Contacts_B_B_C , allConect.Contacts_B_B_D , allConect.Contacts_B_C_C
                        allConect.Contacts_B_C_D])
```
